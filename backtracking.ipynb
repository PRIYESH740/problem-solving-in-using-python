```python
def isValid(row,col,word,board,visited,index, n,m):
    if row <0 or row >= n or col<0 or col >= m or  visited[row][col] or board[row][col]!=word[index]:
        return False
    return True

def constructedWord(row,col,board ,visted,word,k,n,m):
    if k == len(word):
        return True
    drow=[1,-1,0,0]
    dcol=[0,0,1,-1]
    
    for i in range(4):
        next_row = row + drow[i]
        next_col = col + dcol[i]
        
        
        if isValid(next_row,next_col,word,board,visited,k,n,m):
            visited[next_row][next_col] = True
            
            ans= constructedWord(next_row,next_col,board,visited,word,k+1,n,m)
            if ans:
                return True
            
            visited[next_row][next_col] = False
        
    return False



def present(board: List[List[str]], word: str, n: int, m: int) -> bool:
    
    for i in range(n):
        for j in range(m):
            if board[i][j] == word[0]:
                visited=[[False for i in range(m)]for j in range(n)]
                
                ans= constructedWord(i,j,board,visited,word,1,n,m)
                
                if ans:
                    return True

    return False
    
    
```

```python
from typing import List
def dfs(i: int, j: int, n: int, m: int, board: List[List[str]], word: str, k: int) -> bool:
    if k==len(word):
        return True
    elif i<0 or i==n or j<0 or j==m or board[i][j]=="#" or board[i][j]!=word[k]:
        return False
    else:
        temp=board[i][j]

        board[i][j]="#"

        op1=dfs(i+1,j,n,m,board,word,k+1)
        op2=dfs(i,j+1,n,m,board,word,k+1)
        op3=dfs(i-1,j,n,m,board,word,k+1)
        op4=dfs(i,j-1,n,m,board,word,k+1)
        
        board[i][j]=temp

        return op1 or op2 or op3 or op4

def present(board: List[List[str]], word: str, n: int, m: int) -> bool:
    n,m=len(board),len (board[0])
    for i in range(n):
        for j in range(m):
            if board[i][j]==word[0]:
                if dfs(i,j,n,m,board,word,0):
                    return True
    return False
    
```

### Reach the destination

```python
class Solution:
    def ratInAMaze(self, maze):
        path = [[0] * len(maze) for _ in range(len(maze))]
        self.solveMaze(maze, 0, 0, path)
    
    def solveMaze(self, maze, i, j, path):
        n = len(maze)
        if i < 0 or j < 0 or i >= n or j >= n:
            return
        
        if path[i][j] == 1:
            return
        
        if maze[i][j] == 0:
            return
        
        path[i][j] = 1
        
        if i == n - 1 and j == n - 1:
            for row in range(n):
                for col in range(n):
                    print(path[row][col], end=" ")
                print()
            path[i][j] = 0
            return
        
        self.solveMaze(maze, i - 1, j, path)
        self.solveMaze(maze, i, j + 1, path)
        self.solveMaze(maze, i + 1, j, path)
        self.solveMaze(maze, i, j - 1, path)
        path[i][j] = 0
        return
```
