### minimum elements

```python
from os import *
from sys import *
from collections import *
from math import *

from typing import List

def minimumElements(num: List[int], X: int) -> int:
    m = len(num)
    
    # Creating a 2D table to store the subproblem solutions
    table = [[0 for i in range(X + 1)] for i in range(m + 1)]

    # Initialize the first row with a large value (infinity)
    for i in range(1, X + 1):
        table[0][i] = 10**9 +1

    # Dynamic programming to find the minimum number of elements
    
    for i in range(1, m + 1):
        for j in range(1, X + 1):
            if num[i - 1] > j:
                table[i][j] = table[i - 1][j]
            else:
                # Minimum possible for the previous minimum value
                # of the sequence
                table[i][j] = min(table[i - 1][j], table[i][j - num[i - 1]] + 1)
    
    if table[m][X] == 10**9 +1 :
        return -1

    # Return the minimum number of elements needed to make the sum X
    return table[m][X]
    
```

### target sum

```python
def targetSum(arr: List[int], target: int) -> int:
    n = len(arr)
    s = sum(arr)

    # if the target is out of range, return zero
    if target > s or target < -s:
        return 0

    # create a dp array of size (n + 1) x (2 * s + 1)
    dp = [[0] * (2 * s + 1) for _ in range(n + 1)]

    # initialize the base case
    dp[0][s] = 1

    # loop through the array elements
    for i in range(1, n + 1):
        # loop through the possible sums
        for j in range(2 * s + 1):
            # if the current element can be added to the current sum
            if j - arr[i - 1] >= 0:
                # add the number of ways to achieve the same sum by adding the current element
                dp[i][j] += dp[i - 1][j - arr[i - 1]]
            # if the current element can be subtracted from the current sum
            if j + arr[i - 1] < 2 * s + 1:
                # add the number of ways to achieve the same sum by subtracting the current element
                dp[i][j] += dp[i - 1][j + arr[i - 1]]

    # return the final answer
    return dp[n][s + target]
```

### maximum product subarray

```python
def maximumProduct(arr, n):

    if not arr:
        return 0

    n = len(arr)

    # Arrays to store the maximum and minimum product ending at each position
    max_dp = [0] * n
    min_dp = [0] * n

    max_dp[0] = min_dp[0] = arr[0]
    result = arr[0]

    for i in range(1, n):
        if arr[i] < 0:
            max_dp[i] = max(arr[i], min_dp[i - 1] * arr[i])
            min_dp[i] = min(arr[i], max_dp[i - 1] * arr[i])
        else:
            max_dp[i] = max(arr[i], max_dp[i - 1] * arr[i])
            min_dp[i] = min(arr[i], min_dp[i - 1] * arr[i])

        result = max(result, max_dp[i])

    return result
```

### ways to make coin change

```python
def countWaysToMakeChange(denominations, value) :
    dp = [0] * (value + 1)
    dp[0] = 1  # There is one way to make change for value 0 (no coins)

    for coin in denominations:
        for i in range(coin, value + 1):
            dp[i] += dp[i - coin]

    return dp[value]
```

### shortest commom supersequence

```python
from os import *
from sys import *
from collections import *
from math import *

def shortestSupersequence(a: str, b: str) -> str:
	# Write your code here.
    m, n = len(a), len(b)

    # Create a 2D array to store lengths of shortest common supersequences
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the dp array using bottom-up approach
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0 or j == 0:
                dp[i][j] = i + j
            elif a[i - 1] == b[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i - 1][j], dp[i][j - 1])

    # Reconstruct the shortest common supersequence
    i, j = m, n
    supersequence = []
    while i > 0 and j > 0:
        if a[i - 1] == b[j - 1]:
            supersequence.append(a[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] < dp[i][j - 1]:
            supersequence.append(a[i - 1])
            i -= 1
        else:
            supersequence.append(b[j - 1])
            j -= 1

    # Add the remaining characters from a and b
    while i > 0:
        supersequence.append(a[i - 1])
        i -= 1
    while j > 0:
        supersequence.append(b[j - 1])
        j -= 1

    # Reverse the supersequence since we constructed it in reverse order
    supersequence.reverse()

    return ''.join(supersequence)
```
