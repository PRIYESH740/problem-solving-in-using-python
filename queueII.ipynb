### stack using two queue

```python
from sys import stdin
import queue

class Stack :

	#Define data members and __init__()
    def __init__(self):
        self.q1=queue.Queue()
        self.q2=queue.Queue()
        self.count=0
    def getSize(self) :
		#Implement the getSize() function
        return self.count
    def isEmpty(self) :
		#Implement the isEmpty() function
        return self.getSize()==0
    def push(self, data) :
		#Implement the push(element) function
        self.q1.put(data)
        self.count +=1
    def pop(self) :
        if self.count==0 :
            return -1
        while self.q1.qsize()!=1:
            self.q2.put(self.q1.get())
        ans=self.q1.get()
        while self.q2.qsize()!=0:
            self.q1.put(self.q2.get())
        self.count -=1
        return ans 
    def top(self) :
		#Implement the top() function
        if self.count == 0 :
            return -1
        while  self.q1.qsize()!=1:
            self.q2.put(self.q1.get())
        ans=self.q1.get()
        self.q2.put(ans)
        while self.q2.qsize()!=0:
            self.q1.put(self.q2.get())
        return ans
        
```

```python
s=Stack()
s.push(15)
s.push(10)
s.push(20)
print(s.top())
while s.isEmpty() is False:
    print(s.pop())
```
